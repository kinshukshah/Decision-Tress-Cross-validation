```python
import pandas as pd
import numpy as np
```

```python
dt =pd.read_csv('blood.csv')
```

```python
X=dt.iloc[:,0:4]
y=dt.ix[:,4]
```

```python
k=int(input('Enter Value of K'))
```

```python
def kfold(clf,X,y,k): 
    from sklearn.cross_validation import cross_val_score, cross_val_predict
    score  = cross_val_score(clf,X,y,cv=k)
    y_pred =cross_val_predict(clf,X,y,cv=k)
    from sklearn.metrics import accuracy_score
    acc=accuracy_score(y_pred,y)
    return sum(score) / float(len(score))
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
```

```python
maxd=[]
mins=[]
crit=[]
accu=[]
for i in range(4,9):
    for j in range(5,15):
        clf1=tree.DecisionTreeClassifier(criterion='gini',max_depth=i,min_samples_split=j)
        clf2=tree.DecisionTreeClassifier(criterion='entropy',max_depth=i,min_samples_split=j)
        Answer1=kfold(clf1,X,y,k)
        Answer2=kfold(clf2,X,y,k)
        mins.append(j)
        maxd.append(i)
        mins.append(j)
        maxd.append(i)
        crit.append('gini')
        accu.append(Answer1)
        crit.append('entropy')
        accu.append(Answer2)
accuracy_sheet = pd.DataFrame({'Criterion': crit,'Max Depth': maxd,'Minimun Smaple Split': mins,'Accuracy': accu})
```
